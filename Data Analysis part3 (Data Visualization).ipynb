```python
sizes=data.Survived.value_counts()
fig1,ax=plt.subplots()
explode=(0,0.1)
ax.pie(sizes,labels=["Surived","Non-Survived"],autopct="%1.1f%%",shadow=1,explode=explode)
plt.show()
```
